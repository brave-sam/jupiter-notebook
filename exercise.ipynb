## CLASS RESTURANT

```python
class resturant:

    def __init__(self,cuisine,name):

        self.cuisine = cuisine
        self.name = name
        self.number_served =0
    def describe_resturant(self):
        name_food = f"The {self.name} serves {self.cuisine} cuisine  "
        print(name_food)

    def open_resturant(self):
        print(f"The {self.name} is open")
    def number_served_by_resturant(self):
      print(f"This resturant has served {self.number_served} customers")

    def increament_server(self,customers):
      self.number_served = customers

my_resturant = resturant("Turkish","Burak food Resturant")
my_resturant2 = resturant("Italian","Italian food Resturant")
my_resturant3 = resturant("Chinese","OK Noodles food Resturant")
my_resturant.describe_resturant()
my_resturant2.describe_resturant()
my_resturant3.describe_resturant()
my_resturant.open_resturant()
my_resturant2.open_resturant()
my_resturant3.open_resturant()
my_resturant.number_served_by_resturant()
my_resturant.increament_server(509)
my_resturant.number_served_by_resturant()
```

```python
class user:
    def __init__(self,first,last,age,gender,region):
        self.first = first
        self.last = last
        self.age = age
        self.gender = gender
        self.region = region
        self.fullName = first+ " " +last

    def describe_user(self):
        print(f"Name : {self.fullName}")
        print(f"age:{self.age}")
        print(f"gender:{self.gender}")
        print(f"region:{self.region}")



    def greet_user(self):

         print(f"Marhaba ! Hello Glad to know you  {self.fullName}. Your gender is {self.gender} and age {self.age} belongs to {self.region} region")

the_user = user("Sarim","Khan",19,"Male","Asian")

the_user.describe_user()
the_user.greet_user()
```

```python
import random

class Song:

    def __init__(self,song,artist,path):
        self.song = song
        self.artist = artist
        self.path = path

class Playlist:
    def __init__(self):
        self.songs =[]
    def add_song(self,song):
        self.songs.append(song)
    def shuffle(self):
        random.shuffle(self.songs)

class PlayQueue:
    def __init__(self):
        self.queue=[]

    def add_song(self,song):
        self.queue.append(song)

    def play(self):
        if len(self.queue)>0:
            song = self.queue.pop(0)
            print("Playing : ",song)
        else:
            print("The Queue Is empty")

song1=("Song1","Artist:Nishant","location found guilty")
song2=("Song2","Artist:Nishant","location found guilty")
song3=("Song3","Artist:Nishant","location found guilty")
playlist =Playlist()
playlist.add_song(song1)
playlist.add_song(song2)
playlist.add_song(song3)

playlist.shuffle()

play_queue = PlayQueue()
play_queue.add_song(song1)
play_queue.add_song(song2)
play_queue.add_song(song3)

play_queue.play()
play_queue.play()
play_queue.play()

```
