##MAXIMUM NUMBER OF THE LIST


```python
def insertionSort(arr):

    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):

        key = arr[i]

        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >= 0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key


# Driver code to test above
arr = [12, 11, 13, 5, 6]
insertionSort(arr)
for i in range(len(arr)):
    print ("% d" % arr[i])
```

```python
def min_cost(n,red,green,blue):
  costOfPaint = [[0]*3 for _ in range(n)]
  costOfPaint[0]=[red[0],green[0],blue[0]]
  for i in range(1,n):
    costOfPaint[i][0]= red[i]+min(costOfPaint[i-1][1],costOfPaint[i-1][2])
    costOfPaint[i][1]= green[i]+min(costOfPaint[i-1][0],costOfPaint[i-1][2])
    costOfPaint[i][2]= blue[i]+min(costOfPaint[i-1][0],costOfPaint[i-1][1])
  return min(costOfPaint[n-1])
n = int(input("Enter the Number of house: "))
red=[]
green=[]
blue=[]
for i in range(n):
  red.append(int(input("Enter the amount: ")))
  green.append(int(input("Enter the amount: ")))
  blue.append(int(input("Enter the amount: ")))
print(min_cost(n,red,green,blue))
```

```python
def binary_search(n,x):
  low = 0
  high = len(n)-1
  mid = 0
  while low<=high:
    mid = (high + low )//2
    if (n[mid]<x):
      low = mid+1
    elif n[mid]>x:
      high = mid-1
    else:
      return mid
  return -1
list = [1,34,35,36,80]
x = 80
result = binary_search(list,x)

if result != -1:
  print(str(result))
else:
  print("no")


```

## Sieve Of Eratosthenes

```python
def sieve_of_eratosthenes(n):
  primes=[]
  is_prime =[True] * (n+1) #this list will track wheather a number is prime or not

  for p in range(2,n+1):
    if is_prime[p]:#it means p is prime
      primes.append(p) #we are appending the prime in primes list
      for i in range(p*p,n+1,p): #step size of p till n+1
        is_prime[i]=False #It marks all the multiples of p as non prime by setting is_prime = false
  return primes
while True:
  n = int(input("Enter the limit : "))
  print(sieve_of_eratosthenes(n))
  print(f"The total Prime number in between 1 to {n} is: {len(sieve_of_eratosthenes(n))}")
  y= input("You want to continue or not yes/no: ")
  if y =='yes':
    print("Programm Continue...")
    continue
  else:
    print("Terminating the screen...")
    print("Ending the program...")
    break



```
